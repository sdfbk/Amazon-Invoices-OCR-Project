```python
#14 -- Sold By
#44 -- Order  no
#48 --Order Date
#68,69 --Billing Address
#101 -- State ut code
#136 --State
#135 --City
#151 --Place of supply
#157 -- Place of Delivery
#163 --Invoice Number
#169 --Invoice Detail
#48 --Invoice Date
#336 --Tax amount / Total amount
```

```python
amazon_invoices = os.listdir('Amazon')
amazon_invoices
```

```python
sold_by = []
order_no = []
order_date = []
billing_addrs = []
ut_code = []
state = []
city = []
supply = []
delivery = []
invoice_no = []
invoice_detail = []
invoice_date = [] 
for invoice in amazon_invoices:
    invoice = cv2.imread('Amazon/'+ invoice)
    invoice_rgb = cv2.cvtColor(invoice,cv2.COLOR_BGR2RGB)
    df = pytesseract.image_to_data(invoice_rgb,output_type='data.frame')
    sold_by.append(df['text'][14])
    order_no.append(df['text'][44])
    order_date.append(df['text'][48])
    billing_addrs.append(df['text'][68])
    ut_code.append(df['text'][101])
    state.append(df['text'][136])
    city.append(df['text'][135])
    supply.append(df['text'][151])
    delivery.append(df['text'][157])
    invoice_no.append(df['text'][163])
    invoice_detail.append(df['text'][169])
    invoice_date.append(df['text'][48])
```

```python
# piexel values and indexing are different for each image so that's why data will not properly displayed
```

```python
df = pd.DataFrame({"Sold by": sold_by,"Order No":order_no,"Order Date":order_date,"Billing Address":billing_addrs,"State UT Code":ut_code,"State":state,"City":city,"Invoice Supply":supply,"Invoice Delivery":delivery,"Invoice No":invoice_no,"Invoice Detail":invoice_detail,"Invoice Date":invoice_date})
df
```

```python
sold_by = []
order_no = []
order_date = []
billing_addrs = []
ut_code = []
state = []
city = []
supply = []
delivery = []
invoice_no = []
invoice_detail = []
invoice_date = []
tax_amt = []
total_amt = []
invoice_df =  cv2.imread('Amazon/a1.webp')
invoice_df_rgb = cv2.cvtColor(invoice_df,cv2.COLOR_BGR2RGB)
df = pytesseract.image_to_data(invoice_df_rgb,output_type = 'data.frame')
sold_by.append(df['text'][14])
order_no.append(df['text'][44])
order_date.append(df['text'][48])
billing_addrs.append(df['text'][68])
ut_code.append(df['text'][101])
state.append(df['text'][136].split(',')[0])
city.append(df['text'][135].split(',')[0])
supply.append(df['text'][151])
delivery.append(df['text'][157])
invoice_no.append(df['text'][163])
invoice_detail.append(df['text'][169])
invoice_date.append(df['text'][48])
tax_amt.append(df['text'][336].lstrip('3').split(')')[0])
total_amt.append(df['text'][336].lstrip('3').split(')%')[1].rstrip(']'))
```

```python
df = pd.DataFrame({"Sold by": sold_by,"Order No":order_no,"Order Date":order_date,"Billing Address":billing_addrs,"State UT Code":ut_code,"State":state,"City":city,"Invoice Supply":supply,"Invoice Delivery":delivery,"Invoice No":invoice_no,"Invoice Detail":invoice_detail,"Invoice Date":invoice_date,"Tax Amount":tax_amt,"Total Amount":total_amt})
df
```
